```python
# 가장 빠른 거
from heapq import heappop, heappush
import sys
input = sys.stdin.readline
n = int(input())
end_list = []
for _ in range(n):
    a, b = map(int, input().split())
    if a < b:
        end_list.append((b, a))
    else:
        end_list.append((a, b))
d = int(input()) # 철로의 길이
end_list.sort(reverse=True)

contained = []
maxi = 0
while end_list:
    now_end, now_start = end_list.pop()
    now_rail_end = now_end
    now_rail_start = now_rail_end - d

    while contained and contained[0][0] < now_rail_start:
        heappop(contained)

    if now_start >= now_rail_start:
        heappush(contained, (now_start, now_end))

    if maxi < len(contained):
        maxi = len(contained)
print(maxi)
```

```python
# 뻘짓한 거
from heapq import heappop, heappush
import sys
# input = sys.stdin.readline
n = int(input())
start_heap, end_heap = [], []
for _ in range(n):
    a, b = sorted(map(int, input().split()))
    heappush(start_heap, (a, b))
    heappush(end_heap, (b, a))
d = int(input()) # 철로의 길이

contained = [] # 현재 rail_start 기준 담길 수 있는 집-사무실이 담김(heap)
maxi = 0

def set_rail_start(contained=contained, start_heap=start_heap):
    """앞으로 쓸 새로운 레일 시작점 리턴 및 contained 힙 리셋"""
    # 아무것도 contained에 담겨있지 않을 때 작동(초기)
    # print(contained)
    if contained:
        prev_rail_start = heappop(contained)[0] # 지금까지 담긴 애들 중 start가 가장 왼쪽에 있는 애
    while not contained and start_heap:
        start, end = heappop(start_heap)
        if end - start <= d: # 레일 제한에 담을 수 있는 애만
            heappush(contained, (start, end))
            return start
    
    # 이미 contained를 사용해서 뭔가 담겨 있는 경우
    while contained:
        start, end = heappop(contained)
        if prev_rail_start != start:
            return start
    return float('inf')


while start_heap:
    print('hello')
    rail_start = set_rail_start()
    rail_end = rail_start + d

    last_pushed_start = rail_start # 아래 while문에서 포함되는 사람을 못 찾은 경우 대비
    while start_heap:
        start, end = start_heap # 가장 start가 작은 사람
        if start >= rail_start:
            if end <= rail_end: # 담길 수 있는 조건
                last_pushed_start = start
                heappop(start_heap)
                heappush(contained, (start, end))
            else:
                break
        else:
            heappop(start_heap)

    while end_heap:
        end, start = heappop(end_heap) # 가장 end가 작은 사람
        if start <= last_pushed_start:
            continue
        else:
            if end <= rail_end: # 담길 수 있는 조건

                heappush(contained, (start, end))
                last_pushed_end = end
            else:
                break
    print(rail_start, rail_end, start_heap, end_heap, contained)
    if maxi < len(contained):
        maxi = len(contained)
print(maxi)
```
