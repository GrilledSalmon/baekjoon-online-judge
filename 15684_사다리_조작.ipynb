```python
# N : 세로선, M : 가로선, H : 가로선 놓을 수 있는 위치의 개수
import sys
# input = sys.stdin.readline
N, M, H = map(int, input().split())
edge_status = [[0]*(N-1) for _ in range(H)]
for _ in range(M):
    a, b = map(lambda x: int(x)-1, input().split()) # a, b 가로선, 세로선
    edge_status[a][b] = 1

ans = 100
target = list(range(N))

def is_possible(added_bars, res):
    """현재 상태에서 bar를 추가해서 조작을 성공할 수 있는지 확인"""
    diff_cnt = 0
    for i in range(N):
        if target[i] != res[i]:
            diff_cnt += 1
    if diff_cnt > (3-added_bars)*2:
        return False
    return True

def do_ladder(edge_status):
    res = list(range(N))
    for r in range(H):
        for c in range(N-1):
            if edge_status[r][c]:
                res[c], res[c+1] = res[c+1], res[c] # swap
    return res

def is_add_possible(now_row, c):
    """r, c 위치에 bar를 놓을 수 있는지 확인"""
    if now_row[c] == 0:
        if N == 2:
            return True
        if c == 0:
            if now_row[c+1] == 0:
                return True
            return False
        elif c == N-2:
            if now_row[c-1] == 0:
                return True
            return False
        else:
            if now_row[c-1] == 0 and now_row[c+1] == 0:
                return True
            return False
    return False

candi_bars = []
for r in range(H):
    now_row = edge_status[r]
    for c in range(N-1):
        if is_add_possible(now_row, c):
            candi_bars.append((r, c))

def back_tracking(added_bars=0, until_checked=0):
    global ans
    if ans <= added_bars:
        return
    now_res = do_ladder(edge_status)
    if now_res == target: # 조작 성공
        ans = min(added_bars, ans)
        return
    
    if added_bars == 3 or not is_possible(added_bars, now_res):
        return

    for i in range(until_checked, len(candi_bars)):
        r, c = candi_bars[i]
        if is_add_possible(edge_status[r], c):
            edge_status[r][c] = 1 # add
            back_tracking(added_bars+1, i+1)
            edge_status[r][c] = 0 # remove

back_tracking()
if ans == 100:
    print(-1)
else:
    print(ans)
```

```python
# N : 세로선, M : 가로선, H : 가로선 놓을 수 있는 위치의 개수
import sys
# input = sys.stdin.readline
N, M, H = map(int, input().split())
edge_status = [[0]*(N-1) for _ in range(H)]
for _ in range(M):
    a, b = map(lambda x: int(x)-1, input().split()) # a, b 가로선, 세로선
    edge_status[a][b] = 1

ans = 100
target = list(range(N))

def is_possible(added_bars, until_res):
    """현재 상태에서 bar를 추가해서 조작을 성공할 수 있는지 확인"""
    diff_cnt = 0
    for i in range(N):
        if target[i] != until_res[i]:
            diff_cnt += 1
    if diff_cnt > (3-added_bars)*2:
        return False
    return True

def do_ladder(edge_status, until_res:list, start_point):
    res = until_res.copy()
    for r in range(H):
        for c in range(N-1):
            if edge_status[r][c]:
                res[c], res[c+1] = res[c+1], res[c] # swap
    return res

def is_add_possible(now_row, c):
    """r, c 위치에 bar를 놓을 수 있는지 확인"""
    if N == 2:
        return True
    if c == 0:
        if now_row[c+1] == 0:
            return True
        return False
    elif c == N-2:
        if now_row[c-1] == 0:
            return True
        return False
    else:
        if now_row[c-1] == 0 and now_row[c+1] == 0:
            return True
        return False

def back_tracking(added_bars=0, until_res=list(range(N)), start_point=(0, 0)):
    global ans
    if ans == -1:
        return
    now_res = do_ladder(edge_status, until_res, start_point)
    if now_res == target: # 조작 성공
        ans = min(added_bars, ans)
        return

    if added_bars == 3:
        return
    for r in range(H):
        now_row = edge_status[r]
        for c in range(N-1):
            if now_row[c]:
                until_res[c], until_res[c+1] = until_res[c+1], until_res[c]
            elif is_add_possible(now_row, c):
                until_res[c], until_res[c+1] = until_res[c+1], until_res[c]
                edge_status[r][c] = 1
                back_tracking(added_bars+1, until_res, (r, c))
                edge_status[r][c] = 0

back_tracking()
if ans == 100:
    print(-1)
else:
    print(ans)
```
