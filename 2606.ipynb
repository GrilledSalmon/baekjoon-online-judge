- 소감 : 그래프 문제고, BFS, DFS 적혀 있어가지고 쫄았는데 얘는 쉽게 풀 수 있는 문제였음 라고 할 뻔! 사실 완전 틀린 방식으로 풀었다. 그리고 예상치 못한 무한루프가 돌아서 삽질을 많이함.... 원인은 for의 대상 list가 for문 내에서 계속 append가 되고 있기 때문이었음... 물론 list를 잘못 assign해서 같은 레퍼런스를 참조하는지 몰랐음.. 그리고 BFS, DFS 완전 공부 필요... 
- 알고리즘 분류
    - 그래프 이론
    - 그래프 탐색
    - 너비 우선 탐색
    - 깊이 우선 탐색

```python
N = int(input())
M = int(input())

set_1 = set([1])
conn_left = []

conn_list = sorted([sorted(list(map(int, input().split()))) for _ in range(M)])

first_loop = True
while True:
    len_before = len(set_1)
    for conn in conn_list:
        n1, n2 = conn
        if n1 in set_1 or n2 in set_1:
            set_1.update([n1, n2])
        elif first_loop:
            conn_left.append(conn)
    # print(len_before, len(set_1))

    # for문을 돌고 set_1의 변화가 없을 경우(추가로 1과 연결된 요소가 없을 경우)
    if len_before == len(set_1): 
        break
    else: # 처음 loop에서 선택받지 못한 애들 다시 돌려주기(1과 연결될 수 있는 새로운 가능성이 있으므로)
        conn_list = conn_left.copy() # 리스트를 assign하고 for문에서 리스트를 변경하는 경우는 주의한다;;;;
        first_loop = False

print(len(set_1) - 1)
```
