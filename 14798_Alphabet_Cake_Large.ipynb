    - 가로로 확장한 다음에 세로로 확장시키면 무조건 직사각형 모양이 되긴 하는데 이 방법을 떠올리는 게 간단하진 않은 것 같다. 약간 아이큐 테스트 느낌
    - 그리고 일단 문제 이해하는 게 어려워서 한참을 읽다가 현주한테 헬프쳐서 이해함. 계속 영어문제 풀어봐야 할듯
    - 중간중간 인터럽트가 많아서 시간도 오래걸리고 코드도 좀 지저분함
- 소요 시간 : 1시간 20분
- 알고리즘 분류 : 구현, 구성적

```python
# R행 C열의 케잌.
# 각 셀에는 최대 한 개의 이니셜
# 이니셜이 같은 아이는 없음
# 문제가 이해가 안감....
# 같은 글자로 구성된 직사각형 뭉탱이를 만들란 얘기인듯

import sys
# input = sys.stdin.readline

def solve(t):
    R, C = map(int, input().split())
    cake = [list(input().rstrip()) for _ in range(R)]

    def extend_horizontally(r, c):
        "Extend the letter at r,c horizontally"
        letter = cake[r][c]
        # 왼쪽으로
        if c != 0:
            nc = c-1
            while True:
                now_letter = cake[r][nc]
                if now_letter != "?":
                    break
                cake[r][nc] = letter
                if nc == 0:
                    break
                nc -= 1
        
        # 오른쪽으로
        if c != C-1:
            nc = c+1
            while True:
                now_letter = cake[r][nc]
                if now_letter != "?":
                    break
                cake[r][nc] = letter
                if nc == C-1:
                    break
                nc += 1

    def extend_vertically(r, c):
        """Extend the letter at r,c vertically"""
        letter = cake[r][c]
        # 위로
        if r != 0:
            nr = r-1
            while True:
                now_letter = cake[nr][c]
                if now_letter != "?":
                    break
                cake[nr][c] = letter
                if nr == 0:
                    break
                nr -= 1
        
        # 아래로
        if r != R-1:
            nr = r+1
            while True:
                now_letter = cake[nr][c]
                if now_letter != "?":
                    break
                cake[nr][c] = letter
                if nr == R-1:
                    break
                nr += 1

    check_list = []
    for r in range(R):
        for c in range(C):
            if cake[r][c] != '?':
                check_list.append((r, c))

    for r, c in check_list:
        extend_horizontally(r, c)

    # 세로로 확장
    for r in range(R):
        for c in range(C):
            if cake[r][c] != "?":
                extend_vertically(r,c)

    print(f"Case #{t+1}:")

    for line in cake:
        print(''.join(line))

for t in range(int(input())):
    solve(t)
```
