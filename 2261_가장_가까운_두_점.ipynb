- 아이디어
    1. x좌표에 대해 오름차순으로 좌표들을 정렬하고, 정렬한 좌표들을 반씩 쪼갠다.
    2. 2~3개의 점이 남으면 점들 사이의 거리를 구해 최소 거리를 갱신해준다.
    3. left 좌표들과 right 좌표들을 합칠 때 둘 사이의 경계 가까이에 있는 좌표들을 대상으로 band를 만들어 거리를 구해줘야 한다. band를 만드는 기준은 x값이 지금까지의 최소거리이다.
    4. 만들어진 band에 속한 좌표들을 y좌표에 대해 오름차순으로 정렬하여 각 점에 대해 y값 기준 최소거리의 범위 내에서 두 점 사이의 거리를 구해준다.

```python
# 2261 가장 가까운 두 점
import sys
# input = sys.stdin.readline
n = int(input())
coords = sorted([tuple(map(int, input().split())) for _ in range(n)])

mini = float('inf')
mini_root = mini

def dist_square(p, q):
    return (p[0] - q[0])**2 + (p[1] - q[1])**2

def check_mini(p, q):
    """좌표 p,q의 거리가 최소인지 확인하고 저장"""
    global mini, mini_root
    d = dist_square(p, q)
    if d < mini:
        mini = d
        mini_root = mini**0.5

def find_band_coords(start, end):
    """mid로부터의 거리가 mini보다 가까운 좌표들과 x값 기준 리턴"""
    global mini_root
    mid = (start+end)//2
    band_coords = [coords[mid]]
    x_ref = coords[mid][0]
    for i in range(mid-1, start-1, -1): # left(start~mid)
        if abs(coords[i][0] - x_ref) < mini_root:
            band_coords.append(coords[i])
        else: break
    
    for i in range(mid+1, end+1): # right(mid+1~end)
        if abs(coords[i][0] - x_ref) < mini_root:
            band_coords.append(coords[i])
        else:
            break
    band_coords.sort(key= lambda x: x[1])

    return band_coords, x_ref

def merge(start, end):
    """start부터 mid, mid+1부터 end까지 두 그룹에 있는 점들 사이의 거리 중 최솟값 갱신"""
    global mini, mini_root
    band_coords, x_ref = find_band_coords(start, end)
    
    for i in range(len(band_coords)):
        if mini == 0:
            return
        p = band_coords[i]
        if p[0] <= x_ref: # left
            for j in range(1, i): # p의 아래 점들
                q = band_coords[i-j]
                if abs(p[1] - q[1]) < mini_root:
                    if q[0] > x_ref:
                        check_mini(p, q)
                else:
                    break
            for j in range(i+1, len(band_coords)): # p의 윗 점들
                q = band_coords[j]
                if abs(p[1] - q[1]) < mini_root:
                    if q[0] > x_ref:
                        check_mini(p, q)
                else:
                    break
        else: # right
            for j in range(1, i): # p의 아래 점들
                q = band_coords[i-j]
                if abs(p[1] - q[1]) < mini_root:
                    if q[0] <= x_ref:
                        check_mini(p, q)
                else:
                    break
            for j in range(i+1, len(band_coords)): # p의 윗 점들
                q = band_coords[j]
                if abs(p[1] - q[1]) < mini_root:
                    if q[0] <= x_ref:
                        check_mini(p, q)
                else:
                    break

def divide_n_conquer(start=0, end=n-1):
    """start부터 end까지의 점들 사이의 최소 거리를 갱신해줌"""
    if end - start < 3: # 점이 세 개 이하일 때
        for i in range(start, end):
            for j in range(i+1, end+1):
                check_mini(coords[i], coords[j])
        return

    mid = (start + end) // 2
    divide_n_conquer(start, mid)
    divide_n_conquer(mid+1, end)
    merge(start, end)

divide_n_conquer()
print(mini)
```
